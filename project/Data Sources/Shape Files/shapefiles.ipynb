```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

In this notebook, we simply read each of the shapefiles plotting the corresponding boundary maps.

```python
PATH_TO_SHAPEFILES = "./shapefiles/"
```

## Yemen

```python
yemen = gpd.read_file(PATH_TO_SHAPEFILES + "Yemen/Yemen.shp")
yemen.head()
```

```python
fig, ax = plt.subplots(figsize = (20, 7))

ax.set_aspect("equal")
ax.axis("off")
yemen.plot(ax = ax, color = "white", edgecolor = "black")
yemen.centroid.plot(ax = ax, marker = "o", color = "red", markersize = 5)
for x, y, label in zip(yemen.centroid.geometry.x, yemen.centroid.geometry.y, yemen.admin):
    ax.annotate(label, xy = (x, y), xytext = (3, 3), textcoords = "offset points");
```

## Nigeria

```python
nigeria = gpd.read_file(PATH_TO_SHAPEFILES + "Nigeria/Nigeria.shp")
# Not consider entire province of Yobe, Adamawa, Borno because their respective sub-provinces are plotted.
nigeria = nigeria.loc[~nigeria["admin"].isin(["Yobe", "Adamawa", "Borno"])]
nigeria.head()
```

N.B. We create the shapefile of Nigeria adding the sub-provinces of Adamawa, Yobe and Borno provinces (also taking included the entire provinces theirself).

```python
fig, ax = plt.subplots(figsize = (20, 7))

ax.set_aspect("equal")
ax.axis("off")
nigeria.plot(ax = ax, color = "white", edgecolor = "black")
nigeria.centroid.plot(ax = ax, marker = "o", color = "red", markersize = 5)
for x, y, label in zip(nigeria.centroid.geometry.x, nigeria.centroid.geometry.y, nigeria.admin):
    ax.annotate(label, xy = (x, y), xytext = (3, 3), textcoords = "offset points");
```
