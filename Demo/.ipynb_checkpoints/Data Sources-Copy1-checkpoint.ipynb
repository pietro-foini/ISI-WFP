```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("ggplot")
```

# FCS & rCSI & Food Price

```python
# Food Consumption Score (FCS/FCG).
fcs = pd.read_excel("./Food Consumption Score (FCS)/wfp_fcs_yemen.xlsx", index_col = 0, parse_dates = True)

# Reduced Coping Strategy Index (rCSI).
rcsi = pd.read_excel("./Reduced Coping Strategy Index (rCSI)/wfp_rcsi_yemen.xlsx", index_col = 0, parse_dates = True)

# Food Prices.
price = pd.read_excel("./Food Prices (Yemen)/wfp_food_prices_yemen.xlsx", index_col = 0, parse_dates = True)

# Merged.
df = pd.concat([fcs, rcsi, price], axis = 1)
df.head()
```

```python
AdminStratas = set([el[1] for el in df.columns.str.split("_")])
```

```python
for AdminStrata in AdminStratas:
    sub_df = df.loc[:, df.columns.str.endswith(AdminStrata)]
    f = plt.figure(figsize = (11, 5))
    plt.title(AdminStrata, color = "black")
    sub_df.plot(ax = f.gca(), style = ".-")
    plt.legend(title = "Items", loc = "center left", bbox_to_anchor = (1.0, 0.5))
    plt.ylabel("Different metrics")
    plt.xticks(rotation = 90)
    plt.autoscale()
    #plt.savefig(country + '.png')
    plt.show()
```
