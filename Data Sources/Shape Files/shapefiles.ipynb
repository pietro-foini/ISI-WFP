# Shapefiles

The shapefiles are downloaded from OCHA (https://data.humdata.org/) website.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
plt.style.use("default")
```

In this notebook, we simply read each of the shapefiles plotting their corresponding political boundary maps.

```python
PATH_TO_SHAPEFILES = "./shapefiles/"
```

## Yemen

### Administrative level 1

```python
yemen = gpd.read_file(PATH_TO_SHAPEFILES + "Yemen/administrative level 1/Yemen.shp")
yemen.head()
```

```python
fig, ax = plt.subplots(figsize = (20, 7))

ax.set_aspect("equal")
ax.axis("off")
yemen.plot(ax = ax, color = "white", edgecolor = "black")
yemen.centroid.plot(ax = ax, marker = "o", color = "red", markersize = 5)
for x, y, label in zip(yemen.centroid.geometry.x, yemen.centroid.geometry.y, yemen.admin):
    ax.annotate(label, xy = (x, y), xytext = (3, 3), textcoords = "offset points");
```
