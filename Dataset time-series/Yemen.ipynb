# Yemen

I create the dataset of multivariate time-series for the Yemen country.

```python
COUNTRY = "Yemen"
```

```python
PATH_TO_SAVE = "./output_data/Yemen/"
```

## Data sources

For more details regarding each single data source see the folder *Data Sources*.

```python
PATH_TO_DATA_FOLDER = "../Data Sources/"
```

    - Food Consumption Score (FCG <= 2):

```python
# Load the data of the fcs indicator released by wfp.
fcs = pd.read_csv(PATH_TO_DATA_FOLDER + "2. Food Consumption Score (FCS)/output_timeseries/%s/wfp_fcs-interpolate.csv" % COUNTRY, header = [0, 1], index_col = 0)
fcs.index.name = "Datetime"
fcs.index = pd.to_datetime(fcs.index)
freq = "D"
fcs.index.freq = freq
```

```python
fcs
```

    - Reduced Coping Strategy Index (rCSI >= 19):

```python
# Load the the data of the rcsi indicator released by wfp.
rcsi = pd.read_csv(PATH_TO_DATA_FOLDER + "3. Reduced Coping Strategy Index (rCSI)/output_timeseries/%s/wfp_rcsi-interpolate.csv" % COUNTRY, header = [0, 1], index_col = 0)
rcsi.index.name = "Datetime"
rcsi.index = pd.to_datetime(rcsi.index)
freq = "D"
rcsi.index.freq = freq
```

```python
rcsi
```

    - Rainfalls:

```python
# Load the data of rainfall indicator released by wfp.
rainfall = pd.read_csv(PATH_TO_DATA_FOLDER + "4. Rainfall & Vegetation/output_timeseries/%s/wfp_rainfall-dekad.csv" % COUNTRY, header = [0, 1], index_col = 0)
rainfall.index.name = "Datetime"
rainfall.index = pd.to_datetime(rainfall.index)
freq = "D"
rainfall.index.freq = freq
```

```python
rainfall
```

    - Vegetations:

```python
# Load the data of vegetation indicator released by wfp.
NDVI = pd.read_csv(PATH_TO_DATA_FOLDER + "4. Rainfall & Vegetation/output_timeseries/%s/wfp_NDVI-dekad.csv" % COUNTRY, header = [0, 1], index_col = 0)
NDVI.index.name = "Datetime"
NDVI.index = pd.to_datetime(NDVI.index)
freq = "D"
NDVI.index.freq = freq
```

```python
NDVI
```

    - Conflicts:

```python
# Load the data of conflicts indicator released by ACLED.
conflict = pd.read_csv(PATH_TO_DATA_FOLDER + "6. Conflicts/output_timeseries/%s/wfp_conflicts_sum_event_SvyWindow.csv" % COUNTRY, header = [0, 1], index_col = 0)
conflict.index.name = "Datetime"
conflict.index = pd.to_datetime(conflict.index)
freq = "D"
conflict.index.freq = freq
```

```python
conflict
```

    - Ramadan:

```python
# Load the data of Ramadan.
ramadan = pd.read_csv(PATH_TO_DATA_FOLDER + "8. Ramadan/output_timeseries/%s/wfp_ramadan_SvyWindow.csv" % COUNTRY, header = [0, 1], index_col = 0)
ramadan.index.name = "Datetime"
ramadan.index = pd.to_datetime(ramadan.index)
freq = "D"
ramadan.index.freq = freq
```

```python
ramadan
```

    - Adminstrata Codes:

```python
# Load the data of administrative codes.
admin_code = pd.read_csv(PATH_TO_DATA_FOLDER + "7. Adminstrata Code/output_timeseries/%s/wfp_adminstrata_code.csv" % COUNTRY, header = [0, 1], index_col = 0)
admin_code.index.name = "Datetime"
admin_code.index = pd.to_datetime(admin_code.index)
freq = "D"
admin_code.index.freq = freq
```

```python
admin_code
```

    - Population

```python
# Load the data of population released by wfp.
population = pd.read_csv(PATH_TO_DATA_FOLDER + "10. Population/output_timeseries/%s/wfp_population.csv" % COUNTRY, header = [0, 1], index_col = 0)
population.index.name = "Datetime"
population.index = pd.to_datetime(population.index)
freq = "D"
population.index.freq = freq
```

```python
population
```

    - Market Prices

```python
# Load the data of market prices indicator released by wfp.
price = pd.read_csv(PATH_TO_DATA_FOLDER + "5. Market Prices/output_timeseries/%s/wfp_relevant_market_price-interpolate-common-food_group-month.csv" % COUNTRY, header = [0, 1], index_col = 0)
price.index.name = "Datetime"
price.index = pd.to_datetime(price.index)
freq = "D"
price.index.freq = freq
```

```python
price
```

    - Coordinates:

```python
# Load the data of coordination.
coordinates = pd.read_csv(PATH_TO_DATA_FOLDER + "9. Coordinates/output_timeseries/%s/wfp_coordinates.csv" % COUNTRY, header = [0, 1], index_col = 0)
coordinates.index.name = "Datetime"
coordinates.index = pd.to_datetime(coordinates.index)
freq = "D"
coordinates.index.freq = freq
```

```python
coordinates
```

```python
# Create the full dataset.
df = pd.concat([fcs, rcsi, rainfall, conflict, NDVI, admin_code, population, ramadan, price, coordinates], axis = 1, levels = 0).sort_index(axis = 1, level = [0, 1])
```

```python
PROVINCES = df.columns.get_level_values(0).unique()
PROVINCES
```

```python
PREDICTORS = df.columns.get_level_values(1).unique()
PREDICTORS
```

```python
for PROVINCE in PROVINCES:
    print(PROVINCE)
    print(df[PROVINCE].columns)
```

```python
# For now, we decide to delete the Socotra and Hadramaut provinces due to FCG <= 2 and rCSI >= 19 ambiguous behaviour.
df.drop(["Hadramaut", "Socotra"], axis = 1, level = 0, inplace = True)
```

```python
# Plot time-series.
TsIP(df).interactive_plot_df(title = "Time-series", matplotlib = False, style = "mix", comparison = False, first_last_valid_index_group = True)
```

```python
# Save the datasets.
df.to_csv(PATH_TO_SAVE + "Yemen.csv", index_label = False)
```
